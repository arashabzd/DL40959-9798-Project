```
import numpy as np
import matplotlib.pyplot as plt
import torch
from torch import nn, optim
from torch.utils.data import DataLoader
from torchvision import transforms

from model import Model 
import utils
```

```
BATCH_SIZE = 32

corpus_files = ['./Dataset/formulas/train_formulas.txt']
vocab = utils.Vocab(corpus_files)

transform = transforms.Compose([utils.Tokenize(vocab), 
                                utils.Normalize(), 
                                utils.ToTensor()])

trainset = utils.ImageLaTeXDataset(
    image_dir='./Dataset/images/images_train/',
    formulas_file='./Dataset/formulas/train_formulas.txt',
    transform=transform)

trainloader = DataLoader(trainset, 
                         batch_size=BATCH_SIZE, 
                         shuffle=True, 
                         collate_fn=utils.PadSequence(),
                         drop_last=False)


valset = utils.ImageLaTeXDataset(
    image_dir='./Dataset/images/images_validation/',
    formulas_file='./Dataset/formulas/validation_formulas.txt',
    transform=transform)

valloader = DataLoader(valset, 
                       batch_size=BATCH_SIZE, 
                       shuffle=False, 
                       collate_fn=utils.PadSequence(),
                       drop_last=False)
```

```
CUDA = True
encoder_input_dim = 512
decoder_input_dim = vocab.size
hidden_dim = 256
embedding_dim = 80

learning_rate = .001

model = Model(encoder_input_dim, decoder_input_dim, hidden_dim, embedding_dim)
if CUDA:
    model.cuda()

optimizer = optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.CrossEntropyLoss()
```

```
num_epoches = 3
V = 100

for epoch in range(num_epoches):
    
    model.train()
    print('EPOCH {:-3d}:'.format(epoch + 1))
    train_loss = 0.
    for i, (images, formulas) in enumerate(trainloader):
        if CUDA:
            x = images.cuda()
            y = formulas.cuda()
        model.zero_grad()
        y_hat = model(x, y)
        y_hat = torch.transpose(y_hat, 1, 2)
        loss = criterion(y_hat[:, :, :-1], y[:, 1:])
        loss.backward()
        optimizer.step()
        
        train_loss += (loss.item() / V)
        if i % V == (V - 1):
            print('\tIteration {:-4d}: training_loss = {:8.4f}'
                  .format(i + 1, train_loss))
            train_loss = 0.
    
    model.save('./')

    for j in range(10):
        print(vocab.decode(y[j]))
        print(vocab.decode(y_hat[j].argmax(dim=0)))
        plt.imshow(images[j].view(60, 400), cmap='gray')
        plt.show()
    
    val_loss = 0
    with torch.no_grad():
        for j, (val_images, val_formulas) in enumerate(valloader):
            if CUDA:
                x_val = val_images.cuda()
                y_val = val_formulas.cuda()
            y_hat_val = model(x_val, y_val)
            y_hat_val = torch.transpose(y_hat_val, 1, 2)
            loss = criterion(y_hat_val[:, :, :-1], y_val[:, 1:])
            val_loss += loss.item() / len(valloader)
            
    print('validation_loss = {:8.4f}'.format(val_loss))
```

```
valset = utils.ImageLaTeXDataset(image_dir='./Dataset/images/images_validation/',
                                 transform=transform)
valloader = DataLoader(valset, 
                       batch_size=1, 
                       shuffle=False, 
                       drop_last=False)
model.eval()
val_predict = ""
with torch.no_grad():
    for j, (val_images, val_formulas) in enumerate(valloader):
        y_val = torch.LongTensor([vocab.index[vocab.SOS]])
        if CUDA:
            x_val = val_images.cuda()
            y_val = y_val.cuda()
        y_hat_val = model(x_val, y_val)
        y_hat_val = torch.transpose(y_hat_val, 1, 2)
        val_predict += vocab.decode(y_hat_val[0].argmax(dim=0)) + "\n"

with open('./validation_formulas_predict.txt', 'w') as f:
    f.write(val_predict)
```
